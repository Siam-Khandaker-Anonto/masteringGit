# **EDA Function**

```python
def simple_eda(df):
    """Create 6 basic plots"""

    fig, axes = plt.subplots(2, 3, figsize=(12, 8))

    # 1. Math histogram
    axes[0, 0].hist(df['math score'], bins=20)
    axes[0, 0].set_title('Math Score')

    # 2. Gender count
    df['gender'].value_counts().plot(kind='bar', ax=axes[0, 1])
    axes[0, 1].set_title('Gender')

    # 3. Test prep pie
    test = df['test preparation course'].value_counts()
    axes[0, 2].pie(test.values, labels=['None', 'Completed'], autopct='%1.2f%%')
    axes[0, 2].set_title('Test Prep')

    # 4. Math vs Reading
    axes[1, 0].scatter(df['math score'], df['reading score'], alpha=0.3)
    axes[1, 0].set_title('Math vs Reading')

    # 5. Lunch effect
    lunch_avg = df.groupby('lunch')['math score'].mean()
    axes[1, 1].bar(lunch_avg.index, lunch_avg.values)
    axes[1, 1].set_title('Lunch Effect')

    # 6. Average scores
    avg = [df['math score'].mean(), df['reading score'].mean(), df['writing score'].mean()]
    axes[1, 2].bar(['Math', 'Reading', 'Writing'], avg)
    axes[1, 2].set_title('Average Scores')

    plt.tight_layout()
    plt.show()

# Run it
simple_eda(students)
```

# **Gym Data Exploration**

```python
from google.colab import drive
import warnings
warnings.filterwarnings('ignore')
gym = pd.read_csv('/content/drive/MyDrive/datasets/gym_membership.csv')
```

```python
# Gym data

# Display last 5 rows
print(gym.tail())
# Check data types
print(gym.info())
```

```python
import matplotlib.pyplot as plt


# Task 1: Basic Matplotlib (3 plots)
# 1. Bar chart of gym membership types
gym_membership = gym['abonoment_type'].value_counts()

plt.bar(['Standard', 'Premium'], gym_membership.values)
plt.title('Gym Membership Status')
plt.show()

# 2. Histogram of time in gym
gym_time = gym['avg_time_in_gym'].value_counts()

plt.hist(gym_time)
plt.title('Average Time in Gym')
plt.show()

# 3. Scatter plot: Age vs visits per week
age = gym['Age']
visits = gym['visit_per_week']
plt.scatter(age, visits)
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.title('Age VS Visits')
plt.xlabel('Age')
plt.ylabel('Visits')
plt.show()
```

```python
import seaborn as sns

# 1. Box plot: Gender vs Age
sns.boxplot(data=gym, x='gender', y='Age')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title("Gender VS Age")
plt.show()
plt.tight_layout()

# 2. Count plot: Days per week
sns.countplot(data=gym, x='days_per_week')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title("Days Per Week")
plt.show()
plt.tight_layout()

# 3. Bar plot: Average visits by membership type
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title("Avg Visits by Memebership Types")
plt.show()
plt.tight_layout()
```
